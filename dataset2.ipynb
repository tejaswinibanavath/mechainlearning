Interquartile Range (IQR): Measures the statistical dispersion of the data, providing a robust measure of variability. It is calculated as the difference between the third quartile (Q3) and the first quartile (Q1).

Skewness: Measures the asymmetry of the probability distribution of a real-valued random variable. A skewness value of 0 indicates a perfectly symmetrical distribution.

Kurtosis: Measures the tailedness or sharpness of the peak of a distribution. A higher kurtosis value indicates heavier tails and a sharper peak compared to a normal distribution.

Handling Outliers using Z-Score: Identifies outliers by calculating how many standard deviations a data point is from the mean. Points with a z-score greater than a threshold (commonly 3) are considered outliers.

Handling Outliers using IQR: Identifies outliers by comparing data points to the interquartile range. Points outside a certain range (typically 1.5 times IQR) are considered outliers.




```python
import numpy as np
import pandas as pd
from scipy.stats import iqr, skew, kurtosis, zscore
from sklearn.preprocessing import StandardScaler
```

```python
#loading dataset
df = pd.read_csv('/content/Electric_Vehicle_Population_Data.csv')
```

```python
# Extract the numeric columns for analysis
numeric_columns = df.select_dtypes(include=np.number)
```

```python
# Display basic information about the dataset
print(df.info())
```

```python
# Display summary statistics
print(df.describe())
```

```python
# Calculating IQR for each column
Q1 = df.quantile(0.25, numeric_only=True)
Q3 = df.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1

# Identify and remove outliers
outliers = ((df.lt(Q1 - 1.5 * IQR)) | (df.gt(Q3 + 1.5 * IQR))).any(axis=1)
df_no_outliers = df[~outliers]
```

```python
# Print IQR values
print("Q1:\n", Q1)
print("Q3:\n", Q3)
print("IQR:\n", IQR)

# Print outliers
print("Outliers:\n", outliers)

# Print DataFrame without outliers
print("DataFrame without outliers:\n", df_no_outliers)
```

```python
# Calculate skewness and kurtosis for each column
skewness = df.skew()
kurt = df.kurtosis()

print("Skewness:\n", skewness)
print("\nKurtosis:\n", kurt)
```

```python
# Standardize numerical columns using StandardScaler
numerical_columns = df.select_dtypes(include=[np.number]).columns
scaler = StandardScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting boxplots before and after handling outliers
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=df)
plt.title("Boxplot Before Handling Outliers")
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.boxplot(data=df_no_outliers)
plt.title("Boxplot After Handling Outliers")
plt.xticks(rotation=90)

plt.show()
```

```python
# Plotting skewness and kurtosis
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.barplot(x=skewness.index, y=skewness.values)
plt.title("Skewness")
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.barplot(x=kurt.index, y=kurt.values)
plt.title("Kurtosis")

plt.xticks(rotation=45)
plt.show()
```
